```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# Pie Chart

Frekuensi setiap kelas (*class*) divisualisasikan menggunakan lingkaran.

```python
daftar_jenis_kopi_df = pd.DataFrame(
    data = {
        "jenis_kopi": ["latte", "cappucino", "americano", "expresso"],
        "freq": [942, 1716, 731, 220],
        "relative_freq": [0.261, 0.475, 0.203, 0.061],
        "angle": [93.97, 171.17, 72.92, 21.95]
    }
)
daftar_jenis_kopi_df
```

```python
daftar_jenis_kopi_df.plot(
    kind="pie",
    y="relative_freq",
    autopct="%1.1f%%",
    labels=daftar_jenis_kopi_df["jenis_kopi"],
    startangle=90,
    legend=False,
    figsize=(5, 5));
```

# Histogram


- Digunakan pada data bertipe kontinu.
- Frekuensi setiap kelas digambarkan dengan diagram batang. Semakin tinggi diagram batang, maka semakin tinggi juga frekuensi kemunculan kelas tersebut.

```python
df = pd.DataFrame(
    {"x": np.random.normal(0, 0.1, 1000)}
)

df.plot(
    kind="hist",
    legend=False,
    title="Histogram of Normal Distribution"
);
```

# Barplot

- Digunakan pada data bertipe diskrit.
- Frekuensi setiap kelas digambarkan dengan diagram batang. Semakin tinggi diagram batang, maka semakin tinggi juga frekuensi kemunculan kelas tersebut.

```python
daftar_jenis_kopi_df.plot(
    kind="bar",
    x="jenis_kopi",
    y="freq",
    legend=False,
    xlabel="Jenis Kopi",
    ylabel="Frequency");
```

# Scatterplot

- Digunakan untuk data bertipe kontinu.
- Melihat hubungan/relasi antar variabel.

```python
n = 10
x = [i + np.random.randn() for i in np.array(range(n))]
y = [i + np.random.randn() for i in np.array(range(n))]

x_y_df = pd.DataFrame({
    "x": x,
    "y": y
})

x_y_df.plot(
    kind="scatter",
    x="x",
    y="y",
    title="Scatter Plot",
    grid=True
);
```

# Line Chart

- Data divisualisasikan menggunakan garis.
- Setiap titik pada data point merupakan titik temu sumbu x dan y untuk setiap data.

```python
data_penjualan_ayam_geprek_tidak_bebas_df = pd.DataFrame(
    data={
        "dates": ["14-02-2022", "15-02-2022", "16-02-2022", "17-02-2022",
                 "18-02-2022", "19-02-2022", "20-02-2022"],
        "terjual": [100, 230, 250, 200, 260, 280, 300]
    }
)

data_penjualan_ayam_geprek_tidak_bebas_df
```

```python

data_penjualan_ayam_geprek_tidak_bebas_df.plot(
    figsize=(10, 5),
    kind="line",
    x="dates",
    y="terjual",
    xlabel="Waktu (DD-MM-YYYY)",
    ylabel="Terjual (Ekor)",
    legend=False,
    grid=True,
    title="Data Penjualan Ayam Geprek Selama 1 Minggu di Ayam Geprek Tidak Bebas"
);
```
